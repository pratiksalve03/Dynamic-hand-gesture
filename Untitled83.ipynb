# Preprocessing data

```python
desired_frame_count = 45
video_dir = "/content/drive/My Drive/Detection_for_Sign_Language/Include_Data"
categories = os.listdir(video_dir)

for category_index, category in enumerate(categories):
    category_path = os.path.join(video_dir, category)

    if not os.path.isdir(category_path):  # Skip non-folder files
        continue

    print(f"Processing category: {category}")

    category_output_path = os.path.join(category_path, "Extracted_Features")
    os.makedirs(category_output_path, exist_ok=True)

    for video_index, video_file in enumerate(os.listdir(category_path)):
        video_path = os.path.join(category_path, video_file)

        if video_file.endswith(('.mp4', '.avi', '.mkv')):
            continue

        print(f"  Processing video: {video_file}")

        cap = cv2.VideoCapture(video_path)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        frame_count = 0

        if total_frames < desired_frame_count:
            frame_indices = np.linspace(0, total_frames - 1, desired_frame_count).astype(int)
        else:
            frame_indices = np.linspace(0, total_frames - 1, desired_frame_count).astype(int)

        video_output_path = os.path.join(category_output_path, str(video_index))
        os.makedirs(video_output_path, exist_ok=True)

        with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
            for idx in frame_indices:
                cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
                ret, frame = cap.read()
                if not ret:
                    break

                # Process frame
                image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                image.flags.writeable = False
                results = hands.process(image)

                # Extract keypoints
                keypoints = extract_hand_keypoints(results)

                # Save landmarks for the current frame
                frame_feature_file = os.path.join(video_output_path, f"frame_{frame_count}.npy")
                np.save(frame_feature_file, keypoints)
                frame_count += 1

            cap.release()

        print(f"    Saved {frame_count} features for video: {video_file}")
```

# Model Training

```python

model = Sequential()

# Add Bidirectional LSTM layers
model.add(Bidirectional(LSTM(64, return_sequences=True, activation='tanh', input_shape=(45, 126))))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(128, return_sequences=True, activation='tanh')))
model.add(Dropout(0.2))

model.add(LSTM(64, return_sequences=False, activation='tanh'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))

# Add a final Dense layer for classification
model.add(Dense(32, activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax'))  # Adjust 'actions.shape[0]' accordingly

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
import cv2
import numpy as np
import time
from threading import Timer

# Load the pre-trained model
from tensorflow.keras.models import load_model
model = load_model(r'D:\machine learning\SLR\my_model.h5')  # Replace 'my_model.h5' with the actual path to your model file

# Define action labels corresponding to your model's output
actions = ['Accident', 'Call', 'Doctor', 'Help', 'Hot', 'Pain']  # Replace with your gesture labels

# Define colors for visualization
colors = [
    (245, 117, 16),  # Color for 'Accident'
    (117, 245, 16),  # Color for 'Call'
    (16, 117, 245),  # Color for 'Doctor'
    (255, 0, 0),     # Color for 'Help'
    (0, 255, 0),     # Color for 'Hot'
    (0, 0, 255)      # Color for 'Pain'
]

def prob_viz(res, actions, input_frame, colors):
    output_frame = input_frame.copy()
    for num, prob in enumerate(res):
        cv2.rectangle(output_frame, (0, 60 + num * 40), (int(prob * 100), 90 + num * 40), colors[num], -1)
        cv2.putText(output_frame, actions[num], (0, 85 + num * 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
    return output_frame

# 1. New detection variables
sequence = []
sentence = []
predictions = []
threshold = 0.5

# Timer control variables
capture_active = False
capture_duration = 10  # Time to capture frames in seconds
rest_duration = 5      # Rest period in seconds
start_time = time.time()

def toggle_capture():
    global capture_active, start_time
    capture_active = not capture_active
    start_time = time.time()
    Timer(capture_duration if capture_active else rest_duration, toggle_capture).start()

# Start the timer
toggle_capture()

# MediaPipe initialization
import mediapipe as mp
mp_hands = mp.solutions.hands               # Holistic model
mp_drawing = mp.solutions.drawing_utils     # Drawing utilities

def mediapipe_detection(image, model):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_rgb.flags.writeable = False
    results = model.process(image_rgb)
    image_rgb.flags.writeable = True
    return results

def draw_hand_landmarks(image, results):
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(
                image,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                mp_drawing.DrawingSpec(color=(121, 44, 250), thickness=2, circle_radius=2)
            )

def extract_hand_keypoints(results):
    keypoints_right = np.zeros(21 * 3)
    keypoints_left = np.zeros(21 * 3)
    if results.multi_hand_landmarks:
        for hand_index, hand_landmarks in enumerate(results.multi_hand_landmarks):
            handedness = results.multi_handedness[hand_index].classification[0].label
            keypoints = np.array([[lm.x, lm.y, lm.z] for lm in hand_landmarks.landmark]).flatten()
            if handedness == 'Right':
                keypoints_right = keypoints
            elif handedness == 'Left':
                keypoints_left = keypoints
    combined_keypoints = np.concatenate([keypoints_right, keypoints_left])
    return combined_keypoints

cap = cv2.VideoCapture(0)
with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Timer logic
        elapsed_time = int(capture_duration if capture_active else rest_duration - (time.time() - start_time))

        if capture_active:
            # Make detections
            results = mediapipe_detection(frame, hands)
            draw_hand_landmarks(frame, results)
            keypoints = extract_hand_keypoints(results)
            sequence.append(keypoints)
            sequence = sequence[-45:]

            # Prediction logic
            if len(sequence) == 45:
                res = model.predict(np.expand_dims(sequence, axis=0))[0]
                print(actions[np.argmax(res)])
                predictions.append(np.argmax(res))

                # Visualization logic
                if np.unique(predictions[-10:])[0] == np.argmax(res):
                    if res[np.argmax(res)] > threshold:
                        if len(sentence) > 0:
                            if actions[np.argmax(res)] != sentence[-1]:
                                sentence.append(actions[np.argmax(res)])
                        else:
                            sentence.append(actions[np.argmax(res)])

                if len(sentence) > 5:
                    sentence = sentence[-5:]

                frame = prob_viz(res, actions, frame, colors)

        # Display timer and detected sentence
        timer_color = (0, 255, 0) if capture_active else (0, 0, 255)
        cv2.rectangle(frame, (0, 0), (640, 40), timer_color, -1)
        cv2.putText(frame, f'Timer: {elapsed_time}s', (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
        cv2.putText(frame, ' '.join(sentence), (3, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

        cv2.imshow('OpenCV Feed', frame)

        # Break gracefully
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
